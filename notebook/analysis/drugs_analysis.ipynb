```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
drugs = pd.read_csv('../../../data/drugs.csv')
```

```python
drugs.info()
```

```python
drugs['category_id'].value_counts()
```

```python
plt.figure(figsize=(20,10))
sb.countplot(data=drugs, x='category_id')
```

```python
drugs.plot(kind="scatter",x="category_id",y="quantity",c="selling_price",
             s=drugs["selling_price"]/100,label="population",
             figsize=(15,10),cmap=plt.get_cmap("jet"),colorbar=False)
```
